```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 프레임 생성 (독립 변수들)
df = pd.DataFrame({
    'x1': [1, 2, 3, 4, 5],
    'x2': [1, 2, 3, 4, 5],
    'x3': [2, 4, 6, 8, 10],
    'x4': [3, 6, 9, 12, 15]
})

# 상관 행렬 계산
corr_matrix = df.corr()

# 상관 행렬 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```
