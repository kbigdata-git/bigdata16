```python
import numpy as np

# 상태
states = ['s1', 's2', 's3']

# 행동
actions = ['a1', 'a2']

# 보상
rewards = {'s1': {'a1': {'s1': 0, 's2': 0, 's3': 10},
                  'a2': {'s1': 0, 's2': 0, 's3': 0}},
           's2': {'a1': {'s1': 0, 's2': 0, 's3': 0},
                  'a2': {'s1': 0, 's2': 0, 's3': 0}},
           's3': {'a1': {'s1': 0, 's2': 0, 's3': 0},
                  'a2': {'s1': 0, 's2': 0, 's3': 0}}}

# 상태 전이 확률
transitions = {'s1': {'a1': {'s1': 0.5, 's2': 0.5, 's3': 0},
                      'a2': {'s1': 0, 's2': 1, 's3': 0}},
               's2': {'a1': {'s1': 0, 's2': 0, 's3': 1},
                      'a2': {'s1': 0, 's2': 1,'s3': 0}},
               's3': {'a1': {'s1': 0.5, 's2': 0, 's3': 0.5},
                      'a2': {'s1': 0, 's2': 0, 's3': 1}}}

# 감가율
gamma = 0.9

# 가치 함수 초기화
values = {'s1': 0, 's2': 0, 's3': 0}

# 가치 반복 알고리즘
for _ in range(10000):  # enough number of iterations
    new_values = values.copy()
    for state in states:
        for action in actions:
            q_value = 0
            for next_state in states:
                reward = rewards[state][action][next_state]
                transition = transitions[state][action][next_state]
                q_value += transition * (reward + gamma * values[next_state])
            new_values[state] = max(new_values[state], q_value)
    values = new_values

print(values)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 행렬 A를 정의합니다. 이 행렬은 선형 변환을 정의합니다.
A = np.array([[1, 2], [3, 4]])

# 벡터 v를 정의합니다.
v = np.array([5, 6])

# 선형 변환을 적용합니다: Av
Av = np.dot(A, v)

# 원래의 벡터 v를 그립니다.
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')

# 선형 변환을 거친 벡터 Av를 그립니다.
plt.quiver(0, 0, Av[0], Av[1], angles='xy', scale_units='xy', scale=1)

# 그래프의 축을 설정합니다.
plt.xlim(0, 40)
plt.ylim(0, 40)
plt.grid(b=True, which='major')

# 그래프를 표시합니다.
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

# 행렬 A를 정의합니다. 이 행렬은 선형 변환을 정의합니다.
A = np.array([[1, 2], [3, 4]])

# 벡터 v를 정의합니다.
v = np.array([5, 6])

# 그래프의 축을 설정합니다.
ax.set_xlim(0, 40)
ax.set_ylim(0, 40)

quiver = ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')

def update(num):
    Av = np.dot(A, v) * num / 30
    quiver.set_UVC(Av[0], Av[1])

ani = FuncAnimation(fig, update, frames=30, repeat=True)

# plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# 행렬 A를 정의합니다. 이 행렬은 선형 변환을 정의합니다.
A = np.array([[1, 2], [3, 4]])

# 벡터 v를 정의합니다.
v = np.array([5, 6])

fig, ax = plt.subplots()
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.grid(True)

quiver = ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')

def update(num):
    Av = np.dot(A, v) * num / 30
    quiver.set_UVC(Av[0], Av[1])

ani = FuncAnimation(fig, update, frames=30, repeat=True)

plt.show()
```
